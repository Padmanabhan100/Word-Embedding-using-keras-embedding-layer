```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
# sentences

sentences = [
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    "I am a good Developer",
    'understand the meaning of words',
    'I like your videos'
]
```

```python
sentences
```

```python
# initialize the vocabulary size i.e the total number of total possible distinct words present in he whole dataset
voc_size = 10000
```

### One Hot Representation

```python
one_hot_rep = [one_hot(sentence,voc_size) for sentence in sentences]
one_hot_rep
```

* DISTINCT WORD HAVE BEEN ENCODED AS SOME FIXED INDEX

#### WORD EMBEDDING REPRESENTATION

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
import numpy as np
```

```python
max_sent_len = max([len(vec) for vec in one_hot_rep])
embedded = pad_sequences(one_hot_rep,padding='pre',maxlen=max_sent_len)
print(embedded)
```

```python
# Defining dimensions
dimensions = 10
```

```python
# creating a model
model = Sequential()
model.add(Embedding(voc_size,dimensions,input_length=max_sent_len))
model.compile("adam","mse")
```

```python
model.summary()
```

```python
model.predict(embedded)
```
